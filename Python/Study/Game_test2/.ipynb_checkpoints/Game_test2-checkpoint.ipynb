```python
class Action(object):
    
    act_list = ['Mirror', 'Phone', 'TV', 'Todi'] # 모든 행위 목록.
    
    def __init__(self):
        # 역시나 이것도 각 방마다 Override해야함.
        self.act_list = Action.act_list
        self.spec_list = Action.spec_list
        
    def Show_act(self):
        print "이 방에서 할 수 있는 것은"
        print self.act_list
    
    def Work(self):
        self.work = raw_input('>')
        
        if self.work in self.Action_list:
            self.real_work = ActDict.Dict[self.work]
            return self.real_work
        else:
            print "가만히 있습니다."
    
class Room(Action):

    door_list = ['Aroom', 'Broom'] # 일단은 모든 방 목록.
    
    def __init__(self):
        # 여기서는 모든방이지만 각각의 방에서 Override할 예정.
        # class의 variable은 클래스를 instance화 시키지 않아도 사용가능!
        super(Room, self).__init__() # Action에서 __init__을 호출함 = act_list 받아옴.
        self.door_list = Room.door_list 
    
    def Show_room(self):
        print "이동할 수 있는 방은"
        print self.door_list # 어디로 갈 수 있는지 보여줌.
            
    def move(self):
        self.door = raw_input('>') # 일단, 텍스트로 움직일 방을 받음.
        
        if self.door in self.door_list:
            # 텍스트로 받은 방을 Map의 사전으로 방 객체를 호출 (이미 인스턴스화)
            self.next_door = RoomDict.Dict[self.door] 
            
            print "%s으로 이동하였습니다." % self.door
            return self.next_door
        else:
            print "이동하지 않았습니다."
            self.engine = Engine()
            self.engine.statement('%s') % self.__class__.__name__# class name = Room name 이어야 함.
    
    
class Engine(object):
    
    def shuffle(self, x): # self는 클래스의 기본 인스턴스, x는 리스트
        self.x = list(x) # 즉, self.x란 Engine.x를 의미함.
        random.shuffle(self.x)
        return self.x

    def init_point(self): # 시작위치를 랜덤으로 지정할것임.
        self.door_list = Room.door_list
        # Engine 내에 함수가 있기에 self.shuffle을 사용해야함.(self는 기본 인스턴스)
        self.shuffle_list = self.shuffle(self.door_list)
        return self.shuffle_list[0]
    
    def statement(self, room):
        self.room = room
        print "현재위치는 %s 입니다." % self.room
        return RoomDict.Dict[self.room]

class Aroom(Room):
    
    door_list = ['Broom']
    act_list = ['Mirror', 'Phone']
    def __init__(self):
        self.door_list = Aroom.door_list # Room의 init을 Override함.
        self.act_list = Aroom.act_list

class Broom(Room):
    
    door_list = ['Aroom']
    act_list = ['TV', 'Todi']
    def __init__(self):
        self.door_list = Broom.door_list # Room의 init을 Override함.
        self.act_list = Broom.act_list
        
class Mirror(object):
    
    def ment(self):
        print "꺄항"
    
class Phone(object):
    
    def ment(self):
        print "따르릉"
    
class TV(object):
    
    def ment(self):
        print "지잉"

class Todi(object):
    
    def ment(self):
        print "찌끼"
    
    
class RoomDict(object):
    
    a = Aroom()
    b = Broom()
    
    Dict = {
        'Aroom': a,
        'Broom': b
    }
    
class ActDict(object):
    
    m = Mirror()
    p = Phone()
    t = TV()
    T = Todi()
    
    Dict = {
        'Mirror': m,
        'Phone': p,
        'TV': t,
        'Todi':T
    }
    
    Dict2 = {
        'Mirror': '꾸왕!',
        'Phone': '따르릉',
        'TV': '지잉',
        'Todi': '찌끼!'
    }
```
